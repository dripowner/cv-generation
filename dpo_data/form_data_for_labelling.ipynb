```python
import pandas as pd
import json
```

```python
df_clean = pd.read_csv("sft_clean_only_10epochs_from_0_to_999_with_temps.csv")
df_full = pd.read_csv("full_sft_10epochs_from_0_to_999_with_temps.csv")
df_req = pd.read_csv("dpo_requests.csv")
```

```python
df_clean.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
df_full.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
df_req.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
```

```python
df_req = df_req[:1000]
```

```python
df_res = pd.concat([df_req, df_clean, df_full], axis=1)
```

```python
df_res
```

```python
df_res.columns
```

```python
df_res = df_res.reset_index()
```

```python
ans_cols = set(df_res.columns)
ans_cols.discard("index")
ans_cols.discard("source")
list(ans_cols)
```

```python
res = []
for id, row in df_res.iterrows():
    task = {}
    unique_columns = []
    texts = []
    for col in ans_cols:
        if row[col] in texts:
            continue
        texts.append(row[col])
        unique_columns.append(col)
    task["id"] = row["index"]
    task["prompt"] = row["source"]
    task["items"] = [{"title": col, 
                      "body": row[col],
                      "id": col} for col in unique_columns]
    res.append(task)
```

```python
res
```

```python
with open('dpo_for_labelling_0_to_999.json', 'w') as f:
    json.dump(res, f)
```
